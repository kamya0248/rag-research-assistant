```python
from transformers import MarianMTModel, MarianTokenizer

def load_translation_model(src_lang, tgt_lang):
    model_name = f'Helsinki-NLP/opus-mt-{src_lang}-{tgt_lang}'
    tokenizer = MarianTokenizer.from_pretrained(model_name)
    model = MarianMTModel.from_pretrained(model_name)
    return model, tokenizer
```

```python
def translate(text, model, tokenizer):
    inputs = tokenizer(text, return_tensors="pt", padding=True, truncation=True)
    translated = model.generate(**inputs)
    return tokenizer.decode(translated[0], skip_special_tokens=True)
```

```python
# Load models
to_english_model, to_english_tokenizer = load_translation_model("fr", "en")  # French to English
to_french_model, to_french_tokenizer = load_translation_model("en", "fr")  # English to French

# Sample test
input_text = "Quels sont les derniers articles sur l'intelligence artificielle en mÃ©decine ?"
translated_query = translate(input_text, to_english_model, to_english_tokenizer)

print("ðŸ”¤ Translated to English:", translated_query)

# After you get results from semantic search, translate back
summary = "This paper discusses the application of AI in healthcare..."
translated_summary = translate(summary, to_french_model, to_french_tokenizer)

print("ðŸŒ Translated back to French:", translated_summary)
```
